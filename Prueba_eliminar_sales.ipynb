```python
#Importar librerias a utilizar 
import pandas as pd
import numpy as np
```

```python
#Leemos el archivo csv
df = pd.read_csv('compounds_vf.csv')
```

```python
#Vemos la informacion del archivo 
df.info()
```

```python
#Mostramos la estructura del DataFrame
df
```

```python
#Creamos un nuevo DataFrame con la columna smiles y le separamos por el primer . que encuentre en la lista
df1 = df["smiles"].str.split(".", n = 1, expand = True)
df1
```

```python
#Almacenamos el resultado de la separaciòn de la cadena de texto antes del . en un archivo parte1.csv
df1[0].to_csv('parte1.csv')
```

```python
#Almacenamos el resultado de la separaciòn de la cadena de texto despues del . en un archivo parte2.csv
df1[1].to_csv('parte2.csv')
```

```python
#Lee los archivos que antes se han guardado
parte1 = pd.read_csv('parte1.csv')
parte2 = pd.read_csv('parte2.csv')
```

```python
#Unimos los archivos parte1 y parte2 para crear un nuevo dataframe
dataunida=pd.merge(parte1,parte2, on ='Unnamed: 0', how = 'outer')
dataunida
```

```python
#dataunida['1'] = dataunida['1'].fillna(0)
```

```python
dataunida.columns = ['Unnamed: 0','comp0','com1']
```

```python
dataunida['comp0'].astype(str)
dataunida['com1'].astype(str)
dataunida['smile Length0']= dataunida['comp0'].str.len() 
dataunida['smile Length1']= dataunida['com1'].str.len() 
dataunida
```

```python
dataunida['Smile']=np.where((dataunida['smile Length0'] < dataunida['smile Length1']),dataunida['com1'],dataunida['comp0'])
```
